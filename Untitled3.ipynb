### Deaths

```python
def deaths():
    """ Returns Deaths DataFrame"""
    deaths_df_long = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv',
    deaths_df_long = deaths_df.melt(
        id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'],
        value_vars=dates,
        var_name='Date',
        value_name='Deaths'
    )

    deaths_df_long['Date'] = pd.to_datetime(deaths_df_long['Date'])
    deaths_df_long.drop(deaths_df_long[deaths_df_long['Country/Region'].str.contains('Grand Princess') | deaths_df_long['Country/Region'].str.contains('Diamond Princess') | deaths_df_long['Country/Region'].str.contains('MS Zaandam') | deaths_df_long['Province/State'].str.contains('Grand Princess') | deaths_df_long['Province/State'].str.contains('Diamond Princess') | deaths_df_long['Province/State'].str.contains('MS Zaandam')].index,inplace=True)
    # Aggregate the data at Country/Region and Date level.
    deaths_df_long_grouped = deaths_df_long.groupby(['Date','Country/Region'])['Deaths'].sum().reset_index()
    return deaths_df_long_grouped
deaths()
```

```python
def recovered():
    """ Returns Recovered Cases DataFrame"""
    recovered_df_long = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv'
    recovered_df_long = recovered_df.melt(
        id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'],
        value_vars=dates,
        var_name='Date',
        value_name='Recovered'
    )

    recovered_df_long['Date'] = pd.to_datetime(recovered_df_long['Date'])
    recovered_df_long.drop(recovered_df_long[recovered_df_long['Country/Region'].str.contains('Grand Princess') | recovered_df_long['Country/Region'].str.contains('Diamond Princess') | recovered_df_long['Country/Region'].str.contains('MS Zaandam') | recovered_df_long['Province/State'].str.contains('Grand Princess') | recovered_df_long['Province/State'].str.contains('Diamond Princess') | recovered_df_long['Province/State'].str.contains('MS Zaandam')].index,inplace=True)
    
    recovered_df_long_grouped = recovered_df_long.groupby(['Date','Country/Region'])['Recovered'].sum().reset_index()
    return recovered_df_long_grouped
recovered()
```

### Active Cases

### Merge Dataframes

```python
def all_cases():
    """ Merging confirmed and deaths, recovered dataframes
         & Calculating and adding Active Casesdataframe."""

    df = confirmed().merge(right=deaths(),how='left',on=['Country/Region', 'Date'])
    df = df.merge(right=recovered(),how='left',on=['Country/Region','Date'])
    # Creating Active Cases column
    df['Active'] = df['Confirmed'] - (df['Deaths'] + df['Recovered'])
    return df
all_cases()
```

```python
def active():
    """ Returns active cases dataframe"""
    df = all_cases().drop(['Confirmed', 'Deaths', 'Recovered'], axis = 1)
    return df
active()




```

##### Percentage change

```python
def pct_change_weekly():
    """ Calculates Weekly percentage change of all cases DataFrame"""
    df = active().pct_change(freq='5D')
    return df
pct_change_weekly()
    
```
